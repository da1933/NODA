```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, export_graphviz
import graphviz
from sklearn.feature_extraction import DictVectorizer
import sklearn as sklearn
from sklearn.metrics import precision_recall_curve
import project_env as pe
from IPython.display import Image
```

# Import Data

```python
train_dt = pd.read_csv('../data_processing/output/train_dt.csv', encoding = "ISO-8859-1", low_memory=False)
val_dt = pd.read_csv('../data_processing/output/val_dt.csv', encoding = "ISO-8859-1", low_memory=False)
```

```python
X_train_dt, y_train_dt = pe.create_target(train_dt, years = 5)
X_val_dt, y_val_dt = pe.create_target(val_dt, years = 5)
```

```python
#remove non-feature columns
X_train_dt=X_train_dt.drop(['UNIQUE_ID','NEXT_ARREST_TIME','ARREST_DATE_y'], axis=1)
X_val_dt=X_val_dt.drop(['UNIQUE_ID','NEXT_ARREST_TIME','ARREST_DATE_y'], axis=1)
```

```python
X_val_dt.columns
```

# DECISION TREE

```python
'''
Decision tree with entropy criterion
'''

dt = DecisionTreeClassifier(criterion='entropy',max_depth=3)
dt = dt.fit(X_train_dt,y_train_dt)
export_graphviz(dt, feature_names=X_train_dt.columns, out_file='tree_classifier.dot')

pred_train = dt.predict(X_train_dt)
pred_val = dt.predict(X_val_dt)
```

```python
#accuracy

accu_train = sklearn.metrics.accuracy_score(y_train_dt,pred_train)
accu_val = sklearn.metrics.accuracy_score(y_val_dt,pred_val)

print("Accuracy on Training Dataset: {}".format(accu_train))
print("Accuracy on Validation Dataset: {}".format(accu_val))
```

```python
precision = sklearn.metrics.precision_score(y_val_dt,pred_val)
recall = sklearn.metrics.recall_score(y_val_dt,pred_val)
f_score = sklearn.metrics.f1_score(y_val_dt,pred_val)

print("Precision: {}".format(precision))
print("Recall: {}".format(recall))
print("F-Score: {}".format(f_score))
```

```python
#confusion matrix for validation set
cm = sklearn.metrics.confusion_matrix(y_val_dt,pred_val)
cm
```
